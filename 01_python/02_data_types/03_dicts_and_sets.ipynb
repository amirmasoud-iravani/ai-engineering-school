# 2.3 Dictionaries and sets

## Toy ideas

- A **dictionary** is a cabinet with labeled drawers: `key → value`.
- A **set** is a basket that refuses duplicate toys.

## Dictionaries

````python
student = {
    "name": "Mohammad",
    "age": 35,
    "field": "AI engineering",
}

print(student["name"])
print(student.get("age"))
print(student.get("missing_key", "not available"))
````

## Update, add, and delete

````python
student["name"] = "Naghi"
student["last_login"] = "2026-07-11"
del student["age"]
print(student)
````

## Iterate through key/value pairs

````python
user = {"username": "ghader", "role": "admin", "active": True}

for key, value in user.items():
    print(f"{key} -> {value}")
````

## Membership checks dictionary keys by default

````python
print("username" in user)
print("admin" in user.values())
````

## Sets remove duplicates

````python
numbers = [1, 2, 3, 8, 2, 1, 0]
unique_numbers = set(numbers)
print(unique_numbers)
````

Important:

```python
{}      # empty dictionary
set()   # empty set
```

````python
empty_dict = {}
empty_set = set()
print(type(empty_dict).__name__)
print(type(empty_set).__name__)
````

## Add and remove set items

````python
colors = {"red", "green"}
colors.add("white")
colors.remove("red")
colors.discard("red")  # no error when missing
print(colors)
````

## Set operations

````python
a = {"python", "numpy", "pandas"}
b = {"python", "pytorch", "transformers"}

print("union:", a | b)
print("intersection:", a & b)
print("only in a:", a - b)
````

## Frozen sets are immutable

````python
fixed_labels = frozenset({"positive", "negative", "neutral"})
print(fixed_labels)
````

## Tiny exercise: word frequencies

````python
words = ["AI", "Python", "AI", "data", "Python", "AI"]
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

print(frequency)
print("Unique words:", set(words))
````
