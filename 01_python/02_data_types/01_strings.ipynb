# 2.1 Strings

## Toy idea: a string is a necklace of characters

Each character has a position called an **index**. Python starts counting at zero.

```python
text = "Artificial intelligence"
print(text[0])
print(text[1])
print(text[-1])
```

## Slicing: taking a piece of the necklace

```python
print(text[0:10])
print(text[:10])
print(text[11:])
print(text[::2])
```

## Useful string methods

```python
word = "  Python Is Friendly  "
print(word.strip())
print(word.lower())
print(word.upper())
print(word.replace("Friendly", "Powerful"))
```

## Split and join

```python
sentence = "AI_is_the_future"
words = sentence.split("_")
print(words)
print(" ".join(words))
```

## f-strings

```python
name = "Naghi Javadi"
score = 35.38
print(f"{name} has a score of {score:.3f}")
```

## Membership and cleaning

```python
email = "student@example.com"
print("@" in email)

raw = "  hello  "
cleaned = raw.strip()
print(cleaned == "hello")
```

## Counting and locating text

```python
banana = "banana"
print(banana.count("a"))
print(banana.find("na"))   # returns -1 when missing
print(banana.index("na"))  # raises ValueError when missing
```

## Unicode and Persian text

```python
persian_text = "هوش مصنوعی برای پردازش زبان فارسی"
print(persian_text)
print("فارسی" in persian_text)
print(persian_text.split())
```
