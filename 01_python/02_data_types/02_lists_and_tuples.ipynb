# 2.2 Lists and tuples

## Toy idea

- A **list** is a toy train: you may add, remove, and replace carriages.
- A **tuple** is a toy train sealed in a display case: its items cannot be replaced.

```python
cities = ["Tehran", "Yazd", "Shiraz", "Tabriz"]
scores = [16, 20, 18, 17]
mixed = ["AI", 2026, True]

print(cities)
print(scores)
print(mixed)
```

## Indexing and changing a list

```python
users = ["Ali", "Jack", "Sara"]
print(users[0], users[-1])
users[1] = "Jack S."
print(users)
```

## Add items: append, insert, and extend

```python
animals = ["cat", "dog"]
animals.append("parrot")
animals.insert(1, "horse")
animals.extend(["rabbit", "fox"])
print(animals)
```

`append(["rabbit", "fox"])` would add one nested list. `extend(...)` adds each item separately.

```python
example = ["cat"]
example.append(["rabbit", "fox"])
print(example)
```

## Slicing

```python
nums = [0, 10, 20, 30, 40, 50, 60]
print(nums[:3])
print(nums[3:])
print(nums[1:4])
print(nums[-2:])
```

## Remove items

```python
cities = ["Tehran", "Yazd", "Shiraz", "Tabriz"]
cities.remove("Tehran")
del cities[1]
last_city = cities.pop()
print(cities)
print("Popped:", last_city)
```

## Search, count, and sort

```python
letters = ["a", "b", "c", "b"]
print("b" in letters)
print(letters.index("b"))
print(letters.count("b"))

names = ["Ahmad", "hannah", "Sara", "changiz"]
print(sorted(names, key=str.lower))
print("Original is unchanged:", names)
```

## Copying versus sharing the same list

```python
original = [1, 2, 3]
shared = original
copied = original.copy()

original.append(4)
print("shared:", shared)
print("copied:", copied)
```

## Tuples

```python
point = (10, 20)
x, y = point
print(x, y)
```

```python
tokens = ["پردازش", "زبان", "طبیعی", "زبان"]
print("First token:", tokens[0])
print("Unique count:", len(set(tokens)))
print("Number of زبان:", tokens.count("زبان"))
```
