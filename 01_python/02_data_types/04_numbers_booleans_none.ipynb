# 2.4 Numbers, booleans, and None

- `int`: whole numbers.
- `float`: decimal numbers.
- `complex`: numbers with an imaginary part.
- `bool`: `True` or `False`.
- `None`: no value yet / nothing here.

```python
count = 10
accuracy = 0.91
signal = 4j
is_ready = True
result = None

for value in [count, accuracy, signal, is_ready, result]:
    print(value, type(value).__name__)
```

## Converting types

```python
text_number = "42"
number = int(text_number)
print(number + 8)
print(str(number) + " examples")
```
