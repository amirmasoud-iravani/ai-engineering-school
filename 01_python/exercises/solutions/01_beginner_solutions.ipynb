## 2. Strings

```python
text = "  Persian NLP  "
result = text.strip().lower().replace(" ", "_")
print(result)
```

## 3. Lists

```python
tools = ["Python", "Excel", "Git"]
tools.append("PyTorch")
tools.remove("Excel")
tools.sort()
print(tools)
```

## 4. Dictionaries

```python
model = {"name": "baseline", "task": "classification", "score": 0.82}
model["version"] = 1
print(model)
```

## 5. Sets

```python
skills_a = {"Python", "pandas", "Git"}
skills_b = {"Python", "PyTorch", "Git"}
print(skills_a & skills_b)
```

## 6. Operators

```python
accuracy = 0.92
latency_ms = 85
print(accuracy >= 0.90 and latency_ms < 100)
```

## 7. Condition

```python
score = 0.88
if score >= 0.90:
    print("deploy")
else:
    print("improve")
```

## 8. Loop

```python
tokens = ["هوش", "مصنوعی", "فارسی"]
for token in tokens:
    print(token, len(token))
```

## 9. Function

```python
def word_count(text: str) -> int:
    return len(text.split())

print(word_count("پردازش زبان طبیعی فارسی"))
```

## 10. File

```python
from pathlib import Path

path = Path("exercise_text.txt")
path.write_text("هوش مصنوعی\nپردازش زبان\n", encoding="utf-8")
print(path.read_text(encoding="utf-8"))
path.unlink()
```
