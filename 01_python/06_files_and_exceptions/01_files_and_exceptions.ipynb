# 6. Files and exception handling

## Toy ideas

- A file is a notebook stored on the computer.
- An exception is an alarm saying something went wrong.
- `try/except` is a safety net around risky code.

## Work with paths safely

```python
from pathlib import Path

work_dir = Path("example_data")
work_dir.mkdir(exist_ok=True)
text_file = work_dir / "lesson.txt"
print(text_file)
```

## Write, append, and read UTF-8 text

```python
text_file.write_text("This lesson is about files.\n", encoding="utf-8")

with text_file.open("a", encoding="utf-8") as file:
    file.write("Next we learn exception handling.\n")
    file.write("پردازش زبان فارسی\n")

content = text_file.read_text(encoding="utf-8")
print(content)
```

Common modes:

- `r`: read;
- `w`: write and replace old content;
- `a`: append at the end;
- `r+`: read and write an existing file;
- `a+`: append and read;
- `rb` / `wb`: binary files.

`ra` is not a valid mode.

## Catch the error you actually expect

```python
missing_file = work_dir / "missing.txt"

try:
    missing_file.read_text(encoding="utf-8")
except FileNotFoundError:
    print("The file does not exist yet.")
```

## Convert user-like input safely

```python
raw_age = "12a"

try:
    age = int(raw_age)
except ValueError:
    age = None
    print("Age must contain digits only.")

print(age)
```

## try / except / else / finally

```python
def safe_divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Cannot divide by zero.")
        return None
    except TypeError:
        print("Both inputs must be numbers.")
        return None
    else:
        return result
    finally:
        print("Division attempt finished.")

print(safe_divide(8, 2))
print(safe_divide(8, 0))
print(safe_divide("8", 2))
```

## Clean up the example file

```python
text_file.unlink(missing_ok=True)
work_dir.rmdir()
print("Cleaned up.")
```

## Tiny rule

Avoid a bare `except:` unless you truly intend to catch every possible problem. Specific alarms produce clearer code.
