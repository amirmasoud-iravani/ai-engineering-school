# 4. Conditions and loops

## Toy idea

- `if` is a road sign: choose a path.
- a loop is a robot repeating a task.

## if / elif / else

```python
score = 82

if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
else:
    grade = "C"

print(grade)
```

## for loops

```python
tokens = ["Persian", "NLP", "is", "interesting"]
for token in tokens:
    print(token, len(token))
```

## enumerate gives index and value

```python
for index, token in enumerate(tokens):
    print(index, token)
```

## range creates counting values

```python
for number in range(1, 6):
    print(number)
```

## while loops

```python
countdown = 3
while countdown > 0:
    print(countdown)
    countdown -= 1
print("Go!")
```

## break and continue

```python
values = [3, -1, 8, 0, 5]
for value in values:
    if value < 0:
        continue
    if value == 0:
        break
    print(value)
```
