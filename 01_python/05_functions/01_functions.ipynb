# 5. Functions

## Toy idea: a function is a little robot

You give the robot inputs, it performs a job, and it may return an output.

```python
def greet(name):
    """Print a friendly greeting."""
    print(f"Hello, {name}!")


greet("Hassan")
greet("Sara")
```

## Parameters and return values

```python
def add(a, b):
    return a + b

result = add(3, 2)
print(result)
```

## A reusable mean function

```python
def mean(numbers):
    """Return the arithmetic mean of a non-empty collection."""
    if not numbers:
        raise ValueError("numbers must not be empty")
    return sum(numbers) / len(numbers)

print(mean([1, 8, 9]))
print(mean((1, 2, 3)))
```

## Default and keyword arguments

```python
def describe_model(name, task="classification", version=1):
    return f"{name} | task={task} | version={version}"

print(describe_model("baseline"))
print(describe_model("regressor", task="regression", version=2))
```

## Boolean-returning functions

```python
def is_good_movie(imdb_score, threshold=5.0):
    return imdb_score >= threshold

print(is_good_movie(8.3))
print(is_good_movie(3.0))
```

## Type hints

```python
def token_count(text: str) -> int:
    return len(text.split())

print(token_count("Persian natural language processing"))
```

## Scope

```python
message = "outside"

def show_scope():
    message = "inside"
    print(message)

show_scope()
print(message)
```

## Tiny exercise

```python
def normalize_label(label: str) -> str:
    return label.strip().lower().replace(" ", "_")

print(normalize_label("  Positive Comment  "))
```
