# 1. Python basics

## Toy idea: variables are labeled boxes

A variable is a label attached to a value.

````python
name = "Ali"
age = 25
temperature = 18.5

print(name)
print(age)
print(temperature)
````

## Printing several pieces

Use an **f-string** to put values inside a sentence.

````python
temperature = 18
text = f"Temperature is {temperature} degrees centigrade."
print(text)
````

This avoids a common error: Python cannot directly add a string and an integer with `+`.

````python
# Safe demonstration of the error without breaking the notebook:
bad_example = "'Temperature is ' + 18"
print("This expression would raise TypeError:", bad_example)
````

## Reassignment

A label can point to a new value.

````python
age = 23
age = 34
print(age)
````

## Multiple assignment and unpacking

````python
first_name, middle_name, last_name = "Hassan", "Reza", "Alavi"
print(first_name, middle_name, last_name)

scores = [10, 20, 30]
first, second, third = scores
print(first, second, third)
````

## Variable names

Good names may contain letters, digits, and `_`, but they cannot start with a digit. Python is case-sensitive: `name` and `Name` are different labels.

Good:

```python
student_name = "Sara"
score2 = 19
```

Bad:

```python
2score = 19
last name = "Alavi"
%address = "Tehran"
```

````python
student_name = "Sara"
score2 = 19
print(student_name, score2)
````

Python also accepts Unicode identifiers, but English identifiers are usually easier to share across teams.

````python
فایل = "lesson.txt"
print(فایل)
````

## Basic types

````python
examples = {
    "integer": 10,
    "float": 3.14,
    "string": "Laptop",
    "boolean": True,
    "list": [1, 2, 3],
    "tuple": (4, 5, 6),
    "dictionary": {"name": "Ali", "family": "Alavi"},
    "set": {1, 2, 3},
    "none": None,
}

for label, value in examples.items():
    print(f"{label:10} -> {value!r:35} -> {type(value).__name__}")
````
