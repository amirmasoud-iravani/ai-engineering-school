## Importing modules

```python
import math
from collections import Counter

print(math.sqrt(81))
print(Counter(["AI", "NLP", "AI"]))
```

## Standard library versus external package

```python
# math and collections are in the Python standard library.
# NumPy and pandas are external packages installed with pip.
print("Standard-library imports work after Python is installed.")
```

## The main guard

```python
def main():
    print("This is the program entry point.")

if __name__ == "__main__":
    main()
```

## Clean-code habits

- choose descriptive names;
- keep functions small;
- avoid repeating code;
- write docstrings for reusable functions;
- use `pathlib` for paths;
- use UTF-8 explicitly for text files;
- keep secrets and huge datasets out of Git;
- make one meaningful Git commit at a time.

## A small dataclass example

```python
from dataclasses import dataclass

@dataclass
class Experiment:
    name: str
    accuracy: float

run = Experiment(name="baseline", accuracy=0.82)
print(run)
```
