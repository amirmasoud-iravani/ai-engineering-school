# 3. Operators

## Toy idea: operators are buttons

A button tells Python what action to perform on values.

## Arithmetic operators

```python
a = 10
b = 3

print("add:", a + b)
print("subtract:", a - b)
print("multiply:", a * b)
print("divide:", a / b)
print("floor divide:", a // b)
print("remainder:", a % b)
print("power:", a ** b)
```

## Assignment operators

```python
x = 5
x += 1
x *= 2
print(x)
```

## Comparison operators

```python
score = 17
print(score == 17)
print(score != 20)
print(score >= 10)
print(score < 18)
```

## Logical operators

```python
age = 25
has_ticket = True

print(age >= 18 and has_ticket)
print(age < 18 or has_ticket)
print(not has_ticket)
```

## Membership operators

```python
tools = ["Python", "Git", "Jupyter"]
print("Python" in tools)
print("Docker" not in tools)
```

## Equality versus identity

- `==` asks: “Do these objects have equal values?”
- `is` asks: “Are these two labels pointing to the exact same object?”

For normal value comparison, use `==`. A common correct identity check is `value is None`.

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print("a == b:", a == b)
print("a is b:", a is b)
print("a is c:", a is c)
```

## Operator precedence

```python
print(2 + 3 * 4)
print((2 + 3) * 4)
```
