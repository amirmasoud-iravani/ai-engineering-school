## Toy idea

Python is like giving instructions with short, readable sentences. It is popular in data and AI because it is easy to write and connects to powerful libraries.

The course introduction describes Python as:

- **high-level**: you work with friendly ideas instead of tiny machine details;
- **interpreted**: you can run instructions interactively;
- **general-purpose**: it can be used for data, automation, web tools, testing, and AI;
- **easy to learn and use**: the syntax is designed to be readable.

Python itself is not always the fastest language for every low-level calculation. In AI, Python often acts like the friendly driver while optimized C/C++/CUDA code inside libraries does heavy work.
