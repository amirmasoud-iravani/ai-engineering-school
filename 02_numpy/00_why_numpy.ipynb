# 0. Why NumPy?

## Toy idea: a Python list is a mixed toy box; a NumPy array is an organized tray

A Python list can hold many different kinds of objects. A NumPy array normally keeps
one data type and stores values in a compact rectangular structure. This makes large
numerical work easier and usually much faster.

**Learning goals**

- understand what NumPy is for;
- compare a list with an array;
- import NumPy using the standard name `np`;
- see the idea of vectorization;
- know when NumPy is and is not the right tool.


```python
import numpy as np

print("NumPy version:", np.__version__)
```

## A list and an array can look similar

The difference becomes important when we calculate with many numbers.


```python
python_list = [1, 2, 3, 4]
numpy_array = np.array([1, 2, 3, 4])

print(type(python_list))
print(type(numpy_array))
print(numpy_array)
```

## One operation, many values

With a list, `* 2` repeats the list. With a NumPy array, `* 2` multiplies every value.
This "do the same operation to the whole array" style is called **vectorization**.


```python
print("List * 2:", python_list * 2)
print("Array * 2:", numpy_array * 2)
```

## The same task without and with NumPy


```python
temperatures_c = [0, 10, 20, 30]
temperatures_f_loop = []

for value in temperatures_c:
    temperatures_f_loop.append(value * 9 / 5 + 32)

temperatures_c_array = np.array(temperatures_c)
temperatures_f_numpy = temperatures_c_array * 9 / 5 + 32

print("Loop result:", temperatures_f_loop)
print("NumPy result:", temperatures_f_numpy)
```

## A small speed demonstration

Timing results depend on the computer, but the pattern is usually clear: NumPy performs
large numerical operations in optimized compiled code instead of a Python loop.


```python
import time

size = 200_000
values_list = list(range(size))
values_array = np.arange(size)

start = time.perf_counter()
list_result = [value * 2 for value in values_list]
list_time = time.perf_counter() - start

start = time.perf_counter()
array_result = values_array * 2
array_time = time.perf_counter() - start

print(f"List comprehension: {list_time:.6f} seconds")
print(f"NumPy operation:   {array_time:.6f} seconds")
print("First three values agree:", list_result[:3] == array_result[:3].tolist())
```

## Where NumPy appears in AI engineering

NumPy arrays are useful for:

- feature matrices: rows are examples and columns are features;
- images: height × width × color channels;
- word or sentence embeddings: one vector per text;
- model predictions and evaluation calculations;
- preparing data before pandas, scikit-learn, or PyTorch.

NumPy is **not** the best tool for every job. For highly irregular mixed data, normal Python
objects or pandas may be clearer. For GPU neural networks and automatic differentiation,
PyTorch tensors are the natural next step.
