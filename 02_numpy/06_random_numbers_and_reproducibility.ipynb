# 6. Random numbers and reproducibility

## Toy idea: a seed is the starting page of a pretend-random story

Computers create pseudo-random numbers. Using the same seed starts the generator from the
same state, which helps us reproduce experiments.

**Learning goals**

- use NumPy's modern `Generator` API;
- generate integers and floating-point values;
- sample from common distributions;
- shuffle and choose items;
- split indices reproducibly.


```python
import numpy as np
```

## Create a generator


```python
rng = np.random.default_rng(seed=42)
print(rng)
```

## Random integers

The upper bound is excluded, like Python's `range`.


```python
dice_rolls = rng.integers(low=1, high=7, size=10)
print(dice_rolls)
```

## Uniform random values between 0 and 1


```python
uniform_values = rng.random(size=(2, 3))
print(uniform_values)
```

## Normal distribution


```python
normal_values = rng.normal(loc=0.0, scale=1.0, size=8)
print(normal_values)
print("approximately centered near:", normal_values.mean())
```

## Choice


```python
labels = np.array(["positive", "neutral", "negative"])
sampled_labels = rng.choice(labels, size=6, replace=True)
print(sampled_labels)
```

Sampling without replacement chooses distinct items.


```python
sample_without_replacement = rng.choice(np.arange(10), size=4, replace=False)
print(sample_without_replacement)
```

## Shuffle and permutation

- `shuffle` changes an array in place.
- `permutation` returns a shuffled copy.


```python
original = np.arange(8)
shuffled_copy = rng.permutation(original)

print("original:", original)
print("permuted copy:", shuffled_copy)
```

```python
in_place = np.arange(8)
rng.shuffle(in_place)
print(in_place)
```

## Reproducibility

Two generators created with the same seed produce the same sequence.


```python
first_rng = np.random.default_rng(123)
second_rng = np.random.default_rng(123)

first_draw = first_rng.integers(0, 100, size=5)
second_draw = second_rng.integers(0, 100, size=5)

print(first_draw)
print(second_draw)
print("equal:", np.array_equal(first_draw, second_draw))
```

## Reproducible train/validation index split


```python
number_of_examples = 10
indices = rng.permutation(number_of_examples)
split_point = 8

train_indices = indices[:split_point]
validation_indices = indices[split_point:]

print("train:", train_indices)
print("validation:", validation_indices)
```

In real machine-learning projects, scikit-learn offers convenient splitting tools. This
example shows the underlying NumPy idea.
