# 5. Aggregations, axes, and statistics

## Toy idea: an aggregation turns many values into a summary

Examples are sum, mean, minimum, maximum, variance, and standard deviation.

**Learning goals**

- compute common summary statistics;
- understand `axis=0` and `axis=1`;
- find positions with `argmin` and `argmax`;
- preserve dimensions with `keepdims=True`;
- handle missing numerical values represented by `NaN`.


```python
import numpy as np
```

```python
data = np.array([
    [1, 2, 3],
    [4, 5, 6],
])

print(data)
print("shape:", data.shape)
```

## Summary over the whole array


```python
print("sum:", np.sum(data))
print("mean:", np.mean(data))
print("min:", np.min(data))
print("max:", np.max(data))
print("product:", np.prod(data))
```

## Understanding axes

A reliable mental model is: **the selected axis disappears after aggregation**.

- `axis=0` removes the row axis, so values are combined down the rows and one result remains
  for each column.
- `axis=1` removes the column axis, so values are combined across each row.


```python
print("sum, axis=0:", data.sum(axis=0))
print("sum, axis=1:", data.sum(axis=1))
```

```python
print("mean of each column:", data.mean(axis=0))
print("mean of each row:", data.mean(axis=1))
```

## `keepdims=True`

Keeping a length-one axis can make later broadcasting easier.


```python
row_means = data.mean(axis=1, keepdims=True)
centered_rows = data - row_means

print("row means shape:", row_means.shape)
print("row means:\n", row_means)
print("centered rows:\n", centered_rows)
```

## Variance and standard deviation

Variance measures average squared distance from the mean. Standard deviation is the square
root of variance and therefore uses the original unit.


```python
values = np.array([1, 2, 3])
print("variance:", np.var(values))
print("standard deviation:", np.std(values))
```

NumPy uses population variance by default (`ddof=0`). For a sample estimate, use `ddof=1`.


```python
print("population variance:", np.var(values, ddof=0))
print("sample variance:", np.var(values, ddof=1))
```

## Positions of minimum and maximum


```python
scores = np.array([12, 19, 7, 18])
print("position of minimum:", np.argmin(scores))
print("position of maximum:", np.argmax(scores))
```

## Percentiles


```python
ages = np.array([18, 20, 21, 25, 30, 45])
print("median:", np.median(ages))
print("25th, 50th, 75th percentiles:", np.percentile(ages, [25, 50, 75]))
```

## Missing values with `NaN`

`NaN` means "not a number" and is commonly used for missing floating-point values.
Normal aggregations propagate it. The `nan...` versions ignore it.


```python
measurements = np.array([10.0, np.nan, 14.0, 16.0])

print("ordinary mean:", np.mean(measurements))
print("NaN-aware mean:", np.nanmean(measurements))
print("missing mask:", np.isnan(measurements))
```
