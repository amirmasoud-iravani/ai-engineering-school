# 1. Arrays, creation, and properties

## Toy idea: an array is a tray with a shape

A one-dimensional array is like one row of boxes. A two-dimensional array is like a table.
A three-dimensional array is like a stack of tables.

**Learning goals**

- create 1D, 2D, and 3D arrays;
- read `ndim`, `shape`, `size`, and `dtype`;
- use `zeros`, `ones`, `full`, `arange`, and `linspace`;
- convert data types safely;
- understand why rectangular shape matters.


```python
import numpy as np
```

## Create an array from a Python list


```python
a = np.array([1, 2, 3])
print(a)
print(type(a))
```

## Four useful questions about every array

- `ndim`: How many axes does it have?
- `shape`: How long is each axis?
- `size`: How many values are there in total?
- `dtype`: What type of value is stored?


```python
print("ndim:", a.ndim)
print("shape:", a.shape)
print("size:", a.size)
print("dtype:", a.dtype)
```

A 1D array with three values has shape `(3,)`, not `(1, 3)`. The comma tells Python that
this is a one-item tuple describing one axis.


```python
houses = np.array([
    [75, 2, 15],
    [120, 3, 7],
    [55, 1, 30],
])

print(houses)
print("shape:", houses.shape)
print("ndim:", houses.ndim)
print("size:", houses.size)
```

Here, each row can represent one house and the columns can represent area, number of rooms,
and age. The shape `(3, 3)` means 3 rows and 3 columns.


```python
tiny_image_batch = np.array([
    [[0, 255], [255, 0]],
    [[255, 255], [0, 0]],
])

print(tiny_image_batch)
print("shape:", tiny_image_batch.shape)
print("ndim:", tiny_image_batch.ndim)
```

## Common array constructors


```python
zeros = np.zeros((2, 3))
ones = np.ones((2, 3), dtype=np.int32)
eights = np.full((2, 3), 8)

print("zeros:\n", zeros)
print("ones:\n", ones)
print("full:\n", eights)
```

## `arange` versus `linspace`

- `arange(start, stop, step)` controls the step and excludes `stop`.
- `linspace(start, stop, count)` controls the number of evenly spaced values and normally
  includes both endpoints.


```python
even_numbers = np.arange(0, 10, 2)
five_points = np.linspace(0, 1, 5)

print(even_numbers)
print(five_points)
```

## Data types

NumPy usually chooses a type automatically. You can request a type or convert later.
A conversion may lose information—for example, converting `3.9` to an integer removes the
decimal part rather than rounding it.


```python
integers = np.array([1, 2, 3], dtype=np.int32)
floats = integers.astype(np.float32)
truncated = np.array([1.2, 3.9]).astype(np.int32)

print(integers, integers.dtype)
print(floats, floats.dtype)
print(truncated, truncated.dtype)
```

## Helpful inspection


```python
print("item size in bytes:", integers.itemsize)
print("total data bytes:", integers.nbytes)
print("calculation check:", integers.size * integers.itemsize)
```

## Rectangular arrays

A normal numerical NumPy array must be rectangular. Rows with different lengths do not form
a clean numeric matrix. Padding variable-length sequences will be discussed in the NLP lesson.


```python
rectangular = np.array([[1, 2], [3, 4]])
print(rectangular)
print(rectangular.shape)
```
